```python
import json
import numpy as np
from collections import namedtuple
import matplotlib.pyplot as plt
#
def heat(theTemp,Ta,theLambda):
    out=-theLambda*(theTemp-Ta)
    return out
#
def midpoint (npts,tend,To,Ta,theLambda):
    dt=tend/npts
    theTemp=np.empty([npts,],np.float64)
    theTemp[0]=To
    theTime=np.empty_like(theTemp)
# estimate first step by forward euler as need two steps to do leapfrog
    theTemp[1] = To + heat(To,Ta,theLambda)*dt
    theTime[1] = dt
# correct first step by estimating the temperature at a half-step
    Th = To + 0.5*(theTemp[1]-To)
    theTemp[1] = To + heat(Th,Ta,theLambda)*dt
    for timeStep in np.arange(2,npts):
        theTime[timeStep]=theTime[timeStep-1] + dt
        theTemp[timeStep] = theTemp[timeStep-2]+\
                            heat(theTemp[timeStep-1],Ta,theLambda)*2.*dt
    return (theTime,theTemp)
tend=4
Ta=10.
To=20.
theLambda=0.8  #units have to be per minute if time in seconds
npts=40
dt=tend/npts

# OUTPOT FROM MIDPOINT
theTime,theTemp=midpoint(npts,tend,To,Ta,theLambda)
#
# EXACT SOLUTION
exactTime=np.linspace(0,tend,npts)
exactTemp=Ta + (To-Ta)*np.exp(-theLambda*exactTime)
```

```python
plt.figure(figsize=(9,7))
plt.plot(theTime,theTemp,'.-r',label='Midpont Solution')
plt.plot(exactTime,exactTemp,'.-k',label='Exact Solution')
plt.grid(linestyle='--')
plt.xlabel('Time (seconds)')
plt.ylabel('Temperature (C)')
plt.title(r'Newtonian Cooling Equation with: T$_{a}$ = ' + str(Ta) + r',  T$_{0}$ =' + str(To) + r',  $\lambda$ = ' + str(theLambda) + r'  and  $\Delta$t = ' + str(np.round(dt,2)))
plt.legend()
plt.xlim([0,2])
plt.ylim([0,50])
```
